```python
import pandas as pd
from path import Path
```

```python
# loading the data
file_path = Path("./Resources/loans_data.csv")
loans_df = pd.read_csv(file_path)
loans_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
loans_data_encoded = loans_df.copy()
loans_data_encoded["month_le"] = le.fit_transform(loans_data_encoded["month"])
loans_data_encoded.head()
```

```python
# Months dictionary
months_num = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12,
}
```

```python
# Months' names encoded using the dictionary values
loans_data_encoded["month_num"] = loans_data_encoded["month"].apply(lambda x: months_num[x])
loans_data_encoded = loans_data_encoded.drop(columns=["month", "month_le"])
loans_data_encoded.head()
```

```python

loans_data_encoded = pd.get_dummies(loans_data_encoded, columns=['education', 'gender'])
loans_data_encoded.head()
```

```python
file_path = Path('./Resources/loans_data_encoded.csv')
loans_data_encoded.to_csv(file_path, index=False)
```
